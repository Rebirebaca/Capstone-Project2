```python

#Aggregated transaction:
#Get a data from Phonepe data for aggregated transaction using GIT Clone
path="C:/Users/jenis/Desktop/New folder/projectp/pulse/data/aggregated/transaction/country/india/state/"    # Path for access data Folder which us download and stored in local.
aggtd_tran_list= os.listdir(path)

#Dictionary to append data which we iterate from json file
col={"States":[], "Years":[], "Quater":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}

#extract data from aggregate trans Json file
for state in aggtd_tran_list:
    states=path+state+"/"
    aggtd_year_list=os.listdir(states)
    
    for year in aggtd_year_list:
        years=states+year+"/"
        aggtd_file_list=os.listdir(years)
        
        for file in aggtd_file_list:
            files=years+file
            data=open(files,"r")

            X=json.load(data)

            for i in X["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                col["Transaction_type"].append(name)
                col["Transaction_count"].append(count)
                col["Transaction_amount"].append(amount)
                col["States"].append(state)
                col["Years"].append(year)
                col["Quater"].append(int(file.strip(".json")))
```

```python
#dataframe:
agg_transaction=pd.DataFrame(col)
```
